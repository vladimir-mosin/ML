# Практика. Группировки, агрегации и трансформации

```python
import pandas as pd
```

>#### Задание 1
Прочитайте файл `df_flights`.

```python
df_flights = pd.read_csv('data/df_flights.csv')
df_flights
```

#### Контекст

Анализируются сведения о задержках рейсов в авиакомпаниях «Победа», «Уральские_авиалинии», «S7», «Nordwind» и «Аэрофлот».

Признаки:

- Рейс — ID рейса (str), 
- Авиакомпания — название авиакомпании (str), 
- Задержка (мин) — продолжительность задержки рейса в минутах (float), 
- Количество пассажиров — число пассажиров на борту (int),
- День недели — день недели (str).

Данные содержат 3000 записей, пропусков в данных нет.

>#### Задание 2
>- При помощи методов `groupby` и `agg` с использованием внешнего словаря `instruction` выясните, сколько рейсов выполнила каждая авиакомпания.
>- Сделайте то же самое при помощи методов `groupby` и `agg` с именованным атрибутом `flights_count`.
>- Сравните выводы ячеек. Какой из них более информативен?

```python
instruction = {'Авиакомпания': 'count'}
df_flights.groupby('Авиакомпания').agg(instruction)
```

```python
df_flights.groupby('Авиакомпания').agg(flights_count = ('Авиакомпания', 'count'))
```

>#### Задание 3
Вычислите среднюю задержку каждой авиакомпании и среднее квадратичное отклонение задержки. Сделйте это:
>- при помощи методов `groupby` и `agg` с использованием внешнего словаря `instruction`,
>- при помощи методов `groupby` и `agg` с именованными атрибутами `delay_mean` и `delay_std`.
>- Сравните информативность результатов.

```python
instruction = {'Задержка (мин)': ['mean', 'std']}
df_flights.groupby('Авиакомпания').agg(instruction)
```

```python
df_flights.groupby('Авиакомпания').agg(delay_mean=('Задержка (мин)', 'mean'),
                                       delay_std =('Задержка (мин)', 'std'))
```

>#### Задание 4
Для каждой авиакомпании выведите: среднюю и максимальную задержки, среднее количество пассажиров и день недели, в который самолеты этой компании выполняют рейсы чаще всего.  Сделайте это:
>
>- при помощи методов `groupby` и `agg` с использованием внешнего словаря `instruction`,
>- при помощи методов `groupby` и `agg` с именованными атрибутами `delay_mean`, `delay_max`, `number_of_passengers_mean`, `day_of_week_mode`.
>- Сравните информативность результатов.
>
>Подсказка: в обоих случаях для дня недели понадобится вспомогательная функция, так как `mode`  не входит в число стандартных агрегирующих функций.

```python
def mode_custom(series):
    return series.mode()

instruction = {
    'Задержка (мин)': ['mean', 'max'],
    'Количество пассажиров': 'mean',
    'День недели': mode_custom
}

df_flights.groupby('Авиакомпания').agg(instruction)
```

```python
df_flights.groupby('Авиакомпания').agg(delay_mean=('Задержка (мин)', 'mean'), 
                                       delay_max=('Задержка (мин)', 'max'), 
                                       number_of_passengers_mean=('Количество пассажиров', 'mean'), 
                                       day_of_week_mode=('День недели', mode_custom))
```

>#### Задание 5
При помощи методов `groupby` и `transform` сформируйте столбец `'Выраженность задержки'`, в котором будет представлена разность между средней задержкой для данной авиакомпании и задержкой конкретного рейса. 

```python
df_flights['Средняя задержка по компании'] = df_flights.groupby('Авиакомпания')['Задержка (мин)'].transform('mean')
df_flights['Выраженность задержки'] = df_flights['Задержка (мин)'] - df_flights['Средняя задержка по компании']

df_flights
```

>#### Задание 6
При помощи методов `groupby` и `transform` выведите все рейсы, для которых `'Выраженность задержки'` больше, чем две сигмы по компании.

```python
k = 2
mask = df_flights['Выраженность задержки'] > k*df_flights.groupby('Авиакомпания')['Задержка (мин)'].transform('std')
df_flights.loc[mask]
```

>#### Задание 4
Для каждого спортмсена выведите: среднюю и максимальную продолжительность тренировки, среднюю энергозатратность, и средний момент начала тренировки.  Сделайте это:

>#### Задание 5
При помощи методов `groupby` и `transform` сформируйте столбец `'Отклонение от обычной продолжительности тренировки'`, в котором будет представлена разность между средней продолжительностью для данного спортсмена и продолжительностью конкретной тренировки. 

```python
df_fitness['Средняя продолжительность по спортсмену'] = df_fitness.groupby('Спортсмен')['Длительность тренировки (мин)'].transform('mean')
df_fitness['Отклонение от обычной продолжительности тренировки'] = df_fitness['Длительность тренировки (мин)'] - df_fitness['Средняя продолжительность по спортсмену']

df_fitness
```

>#### Задание 6
При помощи методов `groupby` и `transform` веведите все тренировки, для которых `'Отклонение от обычной продолжительности тренировки'` больше, чем две сигмы по спортсмену.

```python
k = 2
mask = df_fitness['Отклонение от обычной продолжительности тренировки'] > k*df_fitness.groupby('Спортсмен')['Длительность тренировки (мин)'].transform('std')
df_fitness.loc[mask]
```
