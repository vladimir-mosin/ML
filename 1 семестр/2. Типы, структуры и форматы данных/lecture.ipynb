## § 2. Структуры данных

Структуры данных в Python — это способы организации данных в памяти, которые позволяют управлять информацией: добавлять, удалять, искать или изменять элементы. Они играют важную роль в анализе данных, так как помогают оптимизировать работу кода, повысить его быстродействие и снизить расход памяти. 

### Некоторые встроенные структуры данных в Python

**Списки (list)** —  упорядоченная изменяемая коллекция элементов, которые могут быть разных типов. Элементы хранятся в памяти последовательно, к ним можно обращаться по индексу. Списки динамические — их размер может меняться. 

**Кортежи (tuple)** —  упорядоченная неизменяемая коллекция элементов. После создания нельзя добавить, удалить или изменить элементы кортежа. Могут содержать разные типы данных: целые числа, строки, списки и т. д.

**Словари (dict)** —  неупорядоченная (до Python 3.7) или упорядоченная (с Python 3.7) изменяемая коллекция пар «ключ — значение». Ключи должны быть уникальными и неизменяемыми (например, строки, числа, кортежи), а значения могут быть любого типа. Начиная с Python 3.7, словари сохраняют порядок добавления элементов. 

**Множества (set)** —  неупорядоченная коллекция уникальных элементов. Дубликаты автоматически удаляются. Элементы множества должны быть неизменяемыми. Оптимизированы для быстрой проверки вхождения элементов, а также для операций объединения, пересечения и разности. 

### Не встроенные, но основные структуры данных для Data Science

**Массивы (np.array)** — структура данных, которая поддерживает произвольное число измерений, векторизованные операции и богатый набор математических функций. 

**Серии (pd.Series)** — одномерная структура данных, представляющая собой помеченный массив. Она может хранить данные любого типа: числа, строки, объекты Python и другие типы. 

**Датафреймы (pd.DataFrame)** — двумерная структура данных, которая представляет собой таблицу со строками и столбцами, организованными в виде двумерного массива. 

## § 3. Списки

Списки (lists) — это одна из самых важных и часто используемых структур данных в Python.

````python
names = ['Катя', 'Петя', 'Леша', 'Аня']
numbers = [1, 3, 3.14, round(2**0.5, 2)]
mixed = ['Катя', 3.14, True]
numbers
````

Обратите внимание, что, если применить к списку функцию `type`, то Python идентифицирует результат, относящийся к типу `list`, хотя на самом деле `list` — это не тип данных, а структура, которая может содержать объекты разных типов.

Основные характеристики списков таковы:
- они индексируемы — элементы списка можно перечислять по порядку,
- изменяемы — можно изменять элементы списков, а также добавлять и удалять элементы,
- могут содержать элементы разных типов,
- поддерживают срезы.

### Индексация списков
Индексация списков начинается с 0. То есть, первый элемент имеет индекс 0, второй — индекс 1 и так далее. Последний элемент имеет индекс, равный длине списка плюс 1 (это не очень удобно, но к этому можно привыкнуть). Зато последний элемент имеет также индекс –1, предпоследний — индекс –2 и так далее (а это очень удобно, и привычка к этому возникает очень быстро). 

Доступ к элементам списка осуществляется по индексу. Например:

````python
print('Первое имя:', names[0])
print('Последнее имя:', names[-1])
print('Предпоследнее имя:', names[-2])
````

Для вычисления длины списка используется встроенная функция `len`:

````python
len(names)
````

### Изменение списков
Изменение списков в Python — это операции, позволяющие модифицировать содержимое списка после его создания: можно добавлять, удалять и заменять элементы. Например, заменим первое имя в списке на Виолетту:

````python
names[0] = 'Виолетта'
names
````

Добавим в конец списка еще одно имя. Пусть это будет Ваня:

````python
names.append('Ваня')
names
````

Вставим между Виолеттой и Петей, (то есть, на первое место) имя Слава:

````python
names.insert(1, 'Слава')
names
````

Обратите внимание, что весь «хвост» списка сохранился, но сдвинулся направо (при этом длина списка, естественно увеличилась на одну позицию).

### Срезы списков
Еще одна важная операция, которую часто приходится выполнять со списками — это срезы. Срезы списков в Python — это механизм, который позволяет извлекать подмножество элементов из списка, создавая новый список с выбранными элементами, при этом исходный список остаётся неизменным. Срезы упрощают работу с данными, позволяя быстро получать нужные фрагменты без использования циклов и дополнительных условий.

Базовый синтаксис среза выглядит так:
```pithon
list_name[begin:end:step]
```
Где `begin` — индекс первого элемента, который будет включён в срез (отсчёт начинается с 0);
`end` — индекс элемента, перед которым срез завершится (этот элемент не включается в срез);
`step` — интервал между выбираемыми элементами (по умолчанию равен 1).

Например, в списке имен все элементы списка с начала до третьего элемента, исключая третий элемент (то есть, по сути, все элементы вплоть до второго включительно) получаются как такой срез:

````python
names[:3]
````

Напомним, что индекс Пети — это 2, то есть, третьего элемента в списке действительно нет. В обратную сторону, все элементы списка имен, начиная с третьего и до конца, получаются так:

````python
names[3:]
````

Элементы с первого по четвертый (включая первый, но исключая четвертый) получаются так:

````python
names[1:4]
````

Это основные срезы, которые используются чаще всего. Вообще, со срезами можно вытворять удивительные штуки, которые далеко не всегда являются очевидными (например, если параметры среза отрицательны, то отсчёт фактически будет вестись справа налево, и что при этом получится — заранее сказать очень трудно). Но нам вполне хватит этих трёх типов срезов, и углубляться в эту тему мы не станем.

### Копии списков

Очень важно правильно создавать копии списков, потому что, если делать это небрежно, то можно получить неожиданные результаты, которые противоречат здравому смыслу. Например, возьмём список имен `names` и создадим новый список имен `names_new`, просто присвоив переменной соответствующее значение:

````python
names_new = names
names_new
````

Теперь добавим имя Джо в конец нового списка: 

````python
names_new.append('Джо')
names_new
````

А теперь выведем старый список имен:

````python
names
````

И вот оно, дикое противоречие. Мы же работали с новым объектом (это нам так кажется, на самом деле, мы работали со старым объектом), почему вместе с ним преобразовался старый? Ответ — потому что, выполнив присваивание, мы не создали новый объект: мы создали новую ссылку на старый объект.

За этим нужно следить и не допускать таких ошибок. Правильное решение — использовать метод `copy` (при этом действительно создаётся новый объект). То есть, если мы выполним сначала копирование, а потом добавление, то старый объект не изменится:

````python
names_new_new = names.copy()

names_new_new.append('Дональд')
names_new_new
````

````python
names
````

Теперь Дональд появился в новом списке, но не появился в старом. Правда, в старом остался Джо, которого мы не собирались туда вносить, и теперь его нужно оттуда вычистить, приложив для этого специальные усилия:

````python
names.remove('Джо')
names
````

### Вложенные списки
Вложенные списки — это списки, элементами которых служат списки. Причем, вложенные списки могут иметь разную длину и относится к разным типам. Например:

````python
double_list = [names, numbers]
double_list
````

С точки зрения математики, это полная и окончательная жуть. И это ещё не все. Списки внутри списков сами по себе могут быть вложенными, обладать различными длинами и уровнями вложенности и так далее. Это ещё большая жуть (для математиков, а вот для программистов это как раз нормально). Мы не будем этого делать. Мы будем воспринимать вложенные списки как матрицы, то есть, ограничимся ситуацией, когда:
- уровней вложенности — один,
- длины вложенных списков — равны между собой.

````python
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]
matrix
````

Отображение у вложенных списков, конечно, так себе. На матрицу не похоже (забегая веред, отметим, что это будет исправлено, когда от списков мы перейдем к массивам `np.array`).

### Генераторы списков

Генераторы списков в Python — это компактный способ создания новых списков на основе существующих итерируемых объектов (списков, строк, диапазонов чисел и т. д.). Они позволяют заменить многострочные циклы `for` одной лаконичной строкой кода.

Базовый синтаксис генератора списка:

```python
[expression for element in array_like if condition]
```
Где `expression` — действие, которое выполняется над каждым элементом; `element` — переменная, принимающая значения из итерируемого объекта; `array_like` — объект, по которому можно проходить в цикле (список, строка, range, одномерный массив, столбец датафрейма и т. п.); `condition` (опционально) — фильтр, оставляющий только элементы, удовлетворяющие заданному критерию.

Вот, например, список квадратов всех целых чисел от 0 до 9 включительно при условии, что эти числа не кратны 3:

````python
[x**2 for x in range(10) if x % 3 != 0]
````

А вот аналог этого кода, исполненный при помощи классических операторов `for` и `if`:

````python
s = []
for x in range(10):
    if x % 3 != 0:
        s.append(x**2)
        
s
````

Генераторы списков обладают целым рядом преимуществ:
- Краткость — сокращают код с нескольких строк до одной.
- Производительность — работают на 20–30 % быстрее обычных циклов.
- Читаемость — чётко описывают намерение («создать список, где каждый элемент — это X для Y из Z»).
- Локализация переменных — переменные внутри генератора не «утекают» в глобальную область видимости.

Действие, которое исполняется в генераторе, не обязательно должно быть математическим. Например, так можно получить список с длинами имён из списка имён, который мы использовали выше:

````python
[len(name) for name in names]
````

Для того, чтобы исполнить полный перебор вариантов, нужно изменить синтаксис:
```python
[expression_1 if condition else expression_2  for element in array_like ]
```
Например, чтобы все некратные тройке числа заменить на их квадраты, а все кратные оставить как есть, выполняется следующий код:

````python
[x**2 if x % 3 != 0 else x for x in range(10)]
````

В общем, это отличный способ получения новых списков на основе старых. И, кстати, исходный итерируемый объект при этом не изменяется, так что можно не опасаться неожиданных сюрпризов. Однако, не стоит злоупотреблять сложностью: для логически запутанной генерации лучше использовать обычные циклы — это повысит читаемость кода и понизит вероятность ошибки.

## § 5. Словари

Словарь в Python — это структура данных, которая хранит информацию в виде пар «ключ: значение» («key»-«value»). 

Это аналог бумажных словарей или записных книжек, где каждому ключу соответствует определённое значение. Считается, что связь между словарем в Python и  словарем в нормальном человеческом понимании — прямолинейная и буквальная, и что это один из тех редких случаев в программировании, когда название идеально описывает суть. На самом деле, связь почти прямолинейная и почти буквальная, а название описывает суть почти идеально (почему это так — см. ниже), а если быть точным, то совсем не идеально.  

### Способы создания словарей

Словари можно создавать разными способами. Например, можно просто указать множество пар:

````python
person = {'name': 'Маша', 'age': 21}
person
````

Здесь ключом является строка 'name', а его единственным значением — строка 'Маша'. И точно так же, еще одним ключом является строка 'age', а его единственным значением — целое число 21.

Можно использовать функцию `dict`, применив ее к списку кортежей:

````python
coordinates = dict([('x', 10), ('y', 20), ('z', 30)])
coordinates
````

Здесь ключи — это строки, а значения ключей — целые числа.

По аналогии со списками, можно использовать генератор словарей:

````python
squares = {x: x**2 for x in range(5)}
squares
````

В этом случае и ключами и значениями являются целые числа.

### Ключи и значения

Доступ к элементам словаря осуществляется по ключу:

````python
person['name']
````

````python
coordinates['x']
````

При этом ключ должен быть уникальным. В этом — фундаментальное отличие между обычным человеческим пониманием словаря и словарем в смысле структуры данных: есть разница между человеческим языком и строгой логикой машин (см. выше, о семантике понятия «словарь»).

Например, в жизни ключ — это многозначное понятие (гаечный ключ, родник и т. д.), а в информатике — это жесткая привязка. Компьютер не умеет понимать контекст. Если бы ключ мог ссылаться на десять разных значений одновременно, система не знала бы, что именно вернуть при запросе.

Чтобы разрешить этот конфликт, используются две стратегии, которые превращают  многозначность в машинную уникальность:
- массивы значений,
- составные ключи.

Можно передать в качестве значения ключа не единственное значение, а список значений:

````python
d = {
    'ключ': ['инструмент, при помощи которого открывается (закрывается) замок', 
             'бьющий из земли источник воды',
             'секретная информация для расшифровки закодированных данных', 
             'инструмент для закручивания (откручивания) гаек']
}
d['ключ']
````

Или, если нужно разграничить значения, можно сделать ключ более специфичным, используя не слово «ключ», а кортеж слов:

````python
d = {
    ('ключ', 'слесарное дело'): 'инструмент для закручивания (откручивания) гаек',
    ('ключ', 'природоведение'): 'бьющий из земли источник воды',
    ('ключ', 'криптография'): 'секретная информация для расшифровки закодированных данных',
    ('ключ', 'домоводство'): 'инструмент, при помощи которого открывается (закрывается) замок',
}
d[('ключ', 'криптография')]
````

Главное отличие понятия «словарь» в бытовом и компьютерном смыслах — это разница между пониманием и поиском:

- Для человека словарь — это инструмент понимания. Человек может прочитать все варианты и выбрать нужный по контексту.
- Для машины словарь — это инструмент поиска в памяти. Машине не нужно понимать, ей нужно гарантированно вернуть один результат.

### Словари в Data Science

Мы уже использовали словари выше: например, когда указывали `dtype` при чтении файлов cvs. Вообще, следует понимать, что  словарь — это инструмент для преобразования данных и хранения вспомогательной логики (именно она, вспомогательная логика, использовалась при чтении файлов с указанием `dtype`). 

Преобразования можно реализовывать при помощи методa `map`. Например, если нам нужно преобразовать словесное описание рейтинга в числовое, то тогда нужные нам соответствия (вспомогательную логику преобразования) мы описываем именно в виде словаря и после этого выполняем мэппинг:

````python
df
````

````python
rating_map = {'Top': 10, 'Medium': 5, 'Low': 0}

df['rating'] = df['rating'].map(rating_map)
df
````

## § 6. Сохраниение типов данных

Формат, ориентированный на сохранение типов данных — это формат parquet. Он создан для эффективного хранения табличных данных с сохранением типов, в отличие от csv, где все — просто текст. Это бинарный формат, то есть, данные в этом формате хранятся не в виде читаемого текста, а в компактном двоичном коде. 

Главная фишка: parquet включают схему данных в заголовок файла. В этой схеме все чётко прописано, например: «столбец age — целое число 64 бита», «столбец name — строка в кодировке UTF-8», «столбец is_active — булевый тип» и т. д. При чтении pandas считывает схему и сразу восстанавливает данные в правильных типах.

Сохраняем данные о продажах гаджетов в формат parquet:

````python
df.to_parquet('data/gadget_sales.parquet')
````
