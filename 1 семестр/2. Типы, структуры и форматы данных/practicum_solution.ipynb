## Преобразования типов

>#### Задание 3
При загрузке данных укажите правильные типы для столбцов:
>- user_id должен быть строкой
>- discount должен быть целым числом
>- conversion должен быть логическим выражением.
>
>Почему возникают ошибки? Загрузите хотя бы ID пользователя как строки. С остальными столбцами придется разбираться по-другому.

```python
df = pd.read_csv('data/marketing_campaign.csv', dtype = {'user_id':str, 'discount':int, 'conversion':bool})
df.head(3)
df.info()
```

>#### Задание 4
Столбец pages_viewed должен быть целым числом, но читается как число с плавающей запятой. Найдите все пропущенные значения в этом столбце, заполните их медианным значением при помощи метода `fillna` и преобразуйте столбец к целому типу при помощи метода `astype`.

```python
df['pages_viewed'] = df['pages_viewed'].fillna(df['pages_viewed'].median()).astype(int)
display(df.head(3))
df.info()
```

>#### Задание 5
Используя мэппинг, замените все значения в столбце conversion на корректные строки 'Yes' и 'No'. Затем при помощи генератора списков преобразуйте этот столбец в стобец с логическими значениями.

```python
df['conversion'].unique()
```

```python
conversion_map = {'Yes':True, 'YES':True, 'No':False, '1':True, '0':False}
df['conversion'] = df['conversion'].map(conversion_map)
```

```python
display(df.head(10))
df.info()
```

>#### Задание 6
Аналогично предыдущему заданию, замените некорректные значения в столбце discount и преобразуйте его к целому типу (пропущенное значение замените на 0).

## Генерация признаков

>#### Задание 9
При помощи генератора списков создайте новый бинарный признак логического типа is_mobile (мобильный пользователь) по принципу: если пользователь использует десктопное приложение, он не мобилен.

```python
df = pd.read_parquet('data/marketing_campaign_new.parquet')
display(df.head(3))
df.info()
```

```python
df['is_mobile'] = [False if x == 'Desktop' else True for x in df['os_type']]
display(df.head(3))
df.info()
```

>#### Задание 10
При помощи мэппинга создайте новый числовой признак age_activity, описывающий среднее время сессии для каждой возрастной группы.

```python
age_activity_map = {'18-25':round(df.loc[df['age_group'] == '18-25']['session_duration'].mean(), 2),
                    '26-35':round(df.loc[df['age_group'] == '26-35']['session_duration'].mean(), 2),
                    '36-50':round(df.loc[df['age_group'] == '36-50']['session_duration'].mean(), 2),
                    '50+'  :round(df.loc[df['age_group'] == '50+'  ]['session_duration'].mean(), 2)}
df['age_activity'] = df['age_group'].map(age_activity_map)
display(df.head(3))
df.info()
```
