>#### Задание 9 (бонус к лекции)
Итак, есть три кита визуального разведочного анализа: `relplot`, `displot`, `catplot`. О них мы уже подробно поговорили. Однако  есть и другие методы подобного рода, своеборазные дельфины — это методы seaborn, которые не входят в большую тройку, и тем не менее, бывают полезны в конкретных ситуациях (но это неточно). 
>- `sns.jointplot()` — функция библиотеки Seaborn для визуализации взаимосвязи между двумя переменными. Она создаёт многопанельный график, который одновременно показывает двумерное распределение двух переменных в центральной панели  и 
одномерные распределения каждой переменной по отдельности на панелях по краям (обычно гистограммы или графики плотности).
>- `sns.pairplot()` — функция библиотеки Seaborn для комплексного анализа взаимосвязей между несколькими переменными в датасете. Она создаёт матрицу графиков, где по диагонали отображаются распределения каждой переменной, а вне диагонали  — диаграммы рассеяния или другие графики, показывающие взаимосвязь между парами переменных.
>- `sns.clustermap()` — функция Seaborn для создания иерархически кластеризованной тепловой карты. Она
визуализирует данные в виде тепловой карты, автоматически группирует строки и столбцы на основе их схожести с помощью иерархической кластеризации и добавляет дендрограммы слева и сверху — они показывают, как формировались группы.
>
> И, хотя на скромный взгляд вашего лектора, все это — не более чем баловство, создающее визуальный шум, тем не менее, запустите ячейки ниже и посмотрите на рузультат.

```python
sns.set_context('notebook')
sns.set_style('white')
```

```python
sns.jointplot(data=df, x='screen_time_bed_hours', y='sleep_quality', kind='hex', color='purple')
plt.suptitle('Дельфин 1: jointplot (hexbin)', y = 1.02)
plt.show()

sns.pairplot(df[['sleep_quality', 'course', 'screen_time_bed_hours', 'caffeine_cups']])
plt.suptitle('Дельфин 2: pairplot', y=1.02)
plt.show()

sns.clustermap(df[['sleep_quality', 'course', 'screen_time_bed_hours', 'caffeine_cups']].corr())
plt.suptitle('Дельфин 3: clustermap (корреляции)', y=1.02)
plt.show()
```
